```python
import os
import json
import yaml
import joblib
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.pipeline import Pipeline
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Optional H5 support
import h5py

# =========================
# PATHS
# =========================
DATASET_PATH = r"C:\Users\NXTWAVE\Downloads\Steel Price Forecasting System\Rajya_Sabha_Session_237_AU1254_1.1.csv"
OUTPUT_DIR = r"C:\Users\NXTWAVE\Downloads\Steel Price Forecasting System"

os.makedirs(OUTPUT_DIR, exist_ok=True)

# =========================
# LOAD DATA
# =========================
df = pd.read_csv(DATASET_PATH)

df.columns = df.columns.str.strip()

month_cols = [
    "Apr.", "May", "June", "July", "Aug.", "Sept.",
    "Oct.", "Nov.", "Dec.", "Jan.", "Feb.", "Mar."
]

# =========================
# CLEAN + RESHAPE DATA
# =========================
long_df = df.melt(
    id_vars=["ITEM", "Year"],
    value_vars=month_cols,
    var_name="Month",
    value_name="Price"
)

long_df.dropna(inplace=True)

month_order = {
    "Apr.": 1, "May": 2, "June": 3, "July": 4,
    "Aug.": 5, "Sept.": 6, "Oct.": 7, "Nov.": 8,
    "Dec.": 9, "Jan.": 10, "Feb.": 11, "Mar.": 12
}

long_df["Month_Num"] = long_df["Month"].map(month_order)

long_df["Start_Year"] = long_df["Year"].apply(lambda x: int(str(x).split("-")[0]))

# Financial year adjustment
long_df["Actual_Year"] = long_df.apply(
    lambda row: row["Start_Year"] + 1 if row["Month_Num"] >= 10 else row["Start_Year"],
    axis=1
)

# Encode item
item_encoder = LabelEncoder()
long_df["ITEM_ENCODED"] = item_encoder.fit_transform(long_df["ITEM"])

# =========================
# FEATURES AND TARGET
# =========================
X = long_df[["ITEM_ENCODED", "Start_Year", "Month_Num", "Actual_Year"]]
y = long_df["Price"]

# =========================
# MODEL TRAINING
# =========================
model = Pipeline([
    ("scaler", StandardScaler()),
    ("regressor", RandomForestRegressor(
        n_estimators=300,
        random_state=42,
        max_depth=10
    ))
])

model.fit(X, y)

predictions = model.predict(X)

# =========================
# METRICS
# =========================
mae = mean_absolute_error(y, predictions)
mse = mean_squared_error(y, predictions)
rmse = np.sqrt(mse)
r2 = r2_score(y, predictions)

metrics = {
    "MAE": float(mae),
    "MSE": float(mse),
    "RMSE": float(rmse),
    "R2_SCORE": float(r2)
}

print("Model Performance:")
print(metrics)

# =========================
# SAVE PKL MODEL
# =========================
pkl_path = os.path.join(OUTPUT_DIR, "steel_price_model.pkl")
joblib.dump(model, pkl_path)

encoder_path = os.path.join(OUTPUT_DIR, "steel_item_encoder.pkl")
joblib.dump(item_encoder, encoder_path)

# =========================
# SAVE JSON
# =========================
json_path = os.path.join(OUTPUT_DIR, "steel_model_metrics.json")
with open(json_path, "w") as f:
    json.dump(metrics, f, indent=4)

# =========================
# SAVE YAML
# =========================
yaml_data = {
    "project": "Steel Price Forecasting System",
    "author": "Sagnik Patra",
    "model": "RandomForestRegressor",
    "features": list(X.columns),
    "target": "Price",
    "metrics": metrics,
    "output_files": {
        "pkl_model": pkl_path,
        "encoder": encoder_path,
        "json_metrics": json_path,
        "yaml_config": "steel_model_config.yaml",
        "h5_data": "steel_model_data.h5"
    }
}

yaml_path = os.path.join(OUTPUT_DIR, "steel_model_config.yaml")
with open(yaml_path, "w") as f:
    yaml.dump(yaml_data, f, sort_keys=False)

# =========================
# SAVE H5 FILE
# =========================
h5_path = os.path.join(OUTPUT_DIR, "steel_model_data.h5")

with h5py.File(h5_path, "w") as h5f:
    h5f.create_dataset("X_features", data=X.values)
    h5f.create_dataset("y_actual", data=y.values)
    h5f.create_dataset("y_predicted", data=predictions)

    h5f.attrs["model_name"] = "RandomForestRegressor"
    h5f.attrs["mae"] = mae
    h5f.attrs["mse"] = mse
    h5f.attrs["rmse"] = rmse
    h5f.attrs["r2_score"] = r2

# =========================
# SAVE PREDICTION CSV
# =========================
result_df = long_df.copy()
result_df["Predicted_Price"] = predictions
result_df["Error"] = result_df["Price"] - result_df["Predicted_Price"]

csv_path = os.path.join(OUTPUT_DIR, "steel_price_predictions.csv")
result_df.to_csv(csv_path, index=False)

# =========================
# FINAL OUTPUT
# =========================
print("\nFiles saved successfully:")
print("PKL Model:", pkl_path)
print("Encoder:", encoder_path)
print("JSON Metrics:", json_path)
print("YAML Config:", yaml_path)
print("H5 File:", h5_path)
print("Prediction CSV:", csv_path)
```
