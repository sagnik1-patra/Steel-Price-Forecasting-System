```python
import os
import json
import yaml
import h5py
import joblib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.pipeline import Pipeline
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# =========================
# PATHS
# =========================
DATASET_PATH = r"C:\Users\NXTWAVE\Downloads\Steel Price Forecasting System\Rajya_Sabha_Session_237_AU1254_1.1.csv"
OUTPUT_DIR = r"C:\Users\NXTWAVE\Downloads\Steel Price Forecasting System"

os.makedirs(OUTPUT_DIR, exist_ok=True)

# =========================
# LOAD DATA
# =========================
df = pd.read_csv(DATASET_PATH)
df.columns = df.columns.str.strip()

month_cols = [
    "Apr.", "May", "June", "July", "Aug.", "Sept.",
    "Oct.", "Nov.", "Dec.", "Jan.", "Feb.", "Mar."
]

# =========================
# RESHAPE DATA
# =========================
long_df = df.melt(
    id_vars=["ITEM", "Year"],
    value_vars=month_cols,
    var_name="Month",
    value_name="Price"
)

long_df.dropna(inplace=True)

long_df["Price"] = pd.to_numeric(long_df["Price"], errors="coerce")
long_df.dropna(inplace=True)

month_order = {
    "Apr.": 1, "May": 2, "June": 3, "July": 4,
    "Aug.": 5, "Sept.": 6, "Oct.": 7, "Nov.": 8,
    "Dec.": 9, "Jan.": 10, "Feb.": 11, "Mar.": 12
}

long_df["Month_Num"] = long_df["Month"].map(month_order)
long_df["Start_Year"] = long_df["Year"].apply(lambda x: int(str(x).split("-")[0]))

long_df["Actual_Year"] = long_df.apply(
    lambda row: row["Start_Year"] + 1 if row["Month_Num"] >= 10 else row["Start_Year"],
    axis=1
)

# =========================
# ENCODING
# =========================
encoder = LabelEncoder()
long_df["ITEM_ENCODED"] = encoder.fit_transform(long_df["ITEM"])

# =========================
# FEATURES AND TARGET
# =========================
X = long_df[["ITEM_ENCODED", "Start_Year", "Month_Num", "Actual_Year"]]
y = long_df["Price"]

# =========================
# MODEL
# =========================
model = Pipeline([
    ("scaler", StandardScaler()),
    ("regressor", RandomForestRegressor(
        n_estimators=300,
        max_depth=10,
        random_state=42
    ))
])

model.fit(X, y)
predictions = model.predict(X)

# =========================
# METRICS
# =========================
mae = mean_absolute_error(y, predictions)
mse = mean_squared_error(y, predictions)
rmse = np.sqrt(mse)
r2 = r2_score(y, predictions)
accuracy = r2 * 100

metrics = {
    "MAE": float(mae),
    "MSE": float(mse),
    "RMSE": float(rmse),
    "R2_SCORE": float(r2),
    "ACCURACY_PERCENTAGE": float(accuracy)
}

print("\nMODEL PERFORMANCE")
print(metrics)

# =========================
# RESULT CSV
# =========================
result_df = long_df.copy()
result_df["Actual_Price"] = y.values
result_df["Predicted_Price"] = predictions
result_df["Error"] = result_df["Actual_Price"] - result_df["Predicted_Price"]

result_csv_path = os.path.join(OUTPUT_DIR, "result_csv.csv")
result_df.to_csv(result_csv_path, index=False)

# =========================
# PREDICTION CSV
# =========================
prediction_df = result_df[
    ["ITEM", "Year", "Month", "Actual_Price", "Predicted_Price", "Error"]
]

prediction_csv_path = os.path.join(OUTPUT_DIR, "prediction_csv.csv")
prediction_df.to_csv(prediction_csv_path, index=False)

# =========================
# SAVE MODEL FILES
# =========================
joblib.dump(model, os.path.join(OUTPUT_DIR, "steel_price_model.pkl"))
joblib.dump(encoder, os.path.join(OUTPUT_DIR, "steel_item_encoder.pkl"))

with open(os.path.join(OUTPUT_DIR, "steel_metrics.json"), "w") as f:
    json.dump(metrics, f, indent=4)

yaml_data = {
    "project": "Steel Price Forecasting System",
    "author": "Sagnik Patra",
    "model": "RandomForestRegressor",
    "features": list(X.columns),
    "target": "Price",
    "metrics": metrics
}

with open(os.path.join(OUTPUT_DIR, "steel_config.yaml"), "w") as f:
    yaml.dump(yaml_data, f, sort_keys=False)

with h5py.File(os.path.join(OUTPUT_DIR, "steel_model_data.h5"), "w") as h5f:
    h5f.create_dataset("X_features", data=X.values)
    h5f.create_dataset("y_actual", data=y.values)
    h5f.create_dataset("y_predicted", data=predictions)
    h5f.attrs["accuracy_percentage"] = accuracy
    h5f.attrs["r2_score"] = r2
    h5f.attrs["mae"] = mae
    h5f.attrs["rmse"] = rmse

# =========================
# 1. ACCURACY GRAPH
# =========================
plt.figure(figsize=(7, 5))
plt.bar(["Model Accuracy"], [accuracy])
plt.ylim(0, 100)
plt.ylabel("Accuracy Percentage")
plt.title("Steel Price Forecasting Model Accuracy")
plt.text(0, accuracy + 1, f"{accuracy:.2f}%", ha="center")
accuracy_graph_path = os.path.join(OUTPUT_DIR, "accuracy_graph.png")
plt.savefig(accuracy_graph_path, dpi=300, bbox_inches="tight")
plt.show()

# =========================
# 2. HEATMAP
# =========================
plt.figure(figsize=(9, 6))
corr_data = result_df[
    ["ITEM_ENCODED", "Start_Year", "Month_Num", "Actual_Year", "Actual_Price", "Predicted_Price", "Error"]
]
sns.heatmap(corr_data.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Feature Correlation Heatmap")
heatmap_path = os.path.join(OUTPUT_DIR, "heatmap.png")
plt.savefig(heatmap_path, dpi=300, bbox_inches="tight")
plt.show()

# =========================
# 3. COMPARISON GRAPH
# =========================
plt.figure(figsize=(12, 6))
plt.plot(result_df["Actual_Price"].values, label="Actual Price")
plt.plot(result_df["Predicted_Price"].values, label="Predicted Price")
plt.title("Actual vs Predicted Steel Price Comparison")
plt.xlabel("Record Index")
plt.ylabel("Steel Price")
plt.legend()
comparison_graph_path = os.path.join(OUTPUT_DIR, "comparison_graph.png")
plt.savefig(comparison_graph_path, dpi=300, bbox_inches="tight")
plt.show()

# =========================
# 4. PREDICTION GRAPH
# =========================
plt.figure(figsize=(10, 6))
plt.scatter(result_df["Actual_Price"], result_df["Predicted_Price"])
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Prediction Graph: Actual vs Predicted")
prediction_graph_path = os.path.join(OUTPUT_DIR, "prediction_graph.png")
plt.savefig(prediction_graph_path, dpi=300, bbox_inches="tight")
plt.show()

# =========================
# 5. RESULT GRAPH
# =========================
plt.figure(figsize=(12, 6))
plt.plot(result_df["Error"].values, label="Prediction Error")
plt.axhline(0, linestyle="--")
plt.title("Result Graph: Prediction Error")
plt.xlabel("Record Index")
plt.ylabel("Error")
plt.legend()
result_graph_path = os.path.join(OUTPUT_DIR, "result_graph.png")
plt.savefig(result_graph_path, dpi=300, bbox_inches="tight")
plt.show()

# =========================
# FINAL PRINT
# =========================
print("\nALL FILES SAVED SUCCESSFULLY")

print("\nCSV FILES:")
print(result_csv_path)
print(prediction_csv_path)

print("\nGRAPH FILES:")
print(accuracy_graph_path)
print(heatmap_path)
print(comparison_graph_path)
print(prediction_graph_path)
print(result_graph_path)

print("\nMODEL FILES:")
print(os.path.join(OUTPUT_DIR, "steel_price_model.pkl"))
print(os.path.join(OUTPUT_DIR, "steel_item_encoder.pkl"))
print(os.path.join(OUTPUT_DIR, "steel_model_data.h5"))
print(os.path.join(OUTPUT_DIR, "steel_metrics.json"))
print(os.path.join(OUTPUT_DIR, "steel_config.yaml"))
```
